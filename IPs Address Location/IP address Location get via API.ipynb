```python
import pandas as pd
import requests
import json
import time
```

```python
ip_data = pd.read_csv('IP_list.csv')
```

```python
ip_data.head(3)
```

```python
ip_data.shape
```

```python
ip_list = ip_data.server_ip.to_list()
```

```python
ip_list
```

```python
len(ip_list)
```

```python
type(ip_list[1])
```

```python
dict1 = {"IP_Address":[],"Country":[],"City":[],"Longitude":[],"Latitude":[]}
```

# Using API -1

```python
for i in ip_list:
    api_request = requests.get("https://tools.keycdn.com/geo.json?host="+str(i))
    api = json.loads(api_request.content)
    
    dict1["IP_Address"].append(api['data']['geo']['ip'])
    dict1["Country"].append(api['data']['geo']['country_name'])
    dict1["City"].append(api['data']['geo']['city'])
    dict1["Longitude"].append(api['data']['geo']['longitude'])
    dict1["Latitude"].append(api['data']['geo']['latitude'])
    
    time.sleep(0.5)

outputdata = pd.DataFrame(dict1)
```

```python
outputdata.head()
```

```python
outputdata.to_csv('Output_API_full.csv',index=False)
```
